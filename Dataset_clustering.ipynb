```python
import pandas as pd
from datetime import datetime
```

```python
ruta_criminalidad = r'./Transformados/Dataset_Criminalidad_Provincial.csv'
ruta_inmigracion = r'./Transformados/Dataset_Inmigracion_Provincial.csv'
ruta_inmigracion_proyeccion = r'./Transformados/Proyeccion_Inmigracion_2023_2024.csv'
ruta_edad_media = r'./Transformados/Dataset_Edad_media_provincias.csv'
ruta_ipc = r'./Transformados/IPC_provincia_Trimestre_normalizado.csv'
ruta_poblacion = r'./Transformados/Poblacion_provincia_anio_trimestre.csv'
ruta_gini = r'./Transformados/proyecciones_gini_prov.csv'
ruta_p80p20 = r'./Transformados/proyecciones_p80p20_prov.csv'
ruta_sociedades = r'./Transformados/Sociedades_mercantiles_año_provincia.csv'
ruta_paro = r'./Transformados/Tasas_provincia_anio.csv'
```

```python
def info(df):
    print(df.shape)
    print(df.dtypes)
#End info
```

```python
criminalidad = pd.read_csv(ruta_criminalidad,
                           sep = ';')
criminalidad.Provincia = criminalidad.Provincia.replace({'Balears (Illes)':'Balears, Illes',
                                                       'Rioja (La)':'Rioja, La',
                                                       'Coruña (A)':'Coruña, A',
                                                       'Palmas (Las)':'Palmas, Las'})
print(criminalidad.shape)
print(criminalidad.dtypes)
criminalidad.head()
```

```python
inmigracion = pd.read_csv(ruta_inmigracion,
                           sep = ';')
inmigracion.Trimestre = inmigracion.Trimestre.str[:1].astype('int64')
inmigracion.Provincia = inmigracion.Provincia.str[3:]
print(inmigracion.shape)
print(inmigracion.dtypes)
inmigracion.head()
```

```python
inmigracion_proyeccion = pd.read_csv(ruta_inmigracion_proyeccion,
                           sep = ',')
inmigracion_proyeccion = inmigracion_proyeccion.pivot_table(index=['Provincia', 'Trimestre','Año'], columns='Origen', values='Proyeccion', aggfunc='sum').reset_index()
inmigracion_proyeccion.Trimestre = inmigracion_proyeccion.Trimestre.str[:1].astype('int64')
inmigracion_proyeccion.Provincia = inmigracion_proyeccion.Provincia.str[3:]
print(inmigracion_proyeccion.shape)
print(inmigracion_proyeccion.dtypes)
inmigracion_proyeccion.head()
```

```python
inmigracion = pd.concat([inmigracion, inmigracion_proyeccion])
print(inmigracion.shape)
print(inmigracion.dtypes)
inmigracion.head()
```

```python
edad_media = pd.read_csv(ruta_edad_media,
                         sep = ';')
info(edad_media)
edad_media.head()
```

```python
ipc = pd.read_csv(ruta_ipc,
                  sep = ';')
info(ipc)
ipc.head()
```

```python
poblacion = pd.read_csv(ruta_poblacion,
                        sep = ';')
info(poblacion)
poblacion.head()
```

```python
gini = pd.read_csv(ruta_gini,
                   sep = ',')
gini.drop(columns = ['ccaa_norm'], inplace = True)
gini = gini.melt( id_vars = 'Provincias',
                 value_vars = ['gini_2017','gini_2018','gini_2019','gini_2020','gini_2021','gini_2022','gini_2023','gini_2024'],
                 var_name = 'Anio',
                 value_name = 'Desigualdad_gini')
gini.Anio = gini.Anio.str[-4:].astype('int64')
gini.Desigualdad_gini = gini.Desigualdad_gini.fillna(gini.Desigualdad_gini.mean())
gini.rename(columns = {'Provincias':'Provincia'}, inplace = True)
info(gini)
gini.head()
```

```python
p80p20 = pd.read_csv(ruta_p80p20,
                   sep = ',')
p80p20.head()
```

```python
p80p20 = pd.read_csv(ruta_p80p20,
                   sep = ',')
p80p20.drop(columns = ['ccaa_norm'], inplace = True)
p80p20 = p80p20.melt( id_vars = 'Provincias',
                 value_vars = ['p80p20_2017','p80p20_2018','p80p20_2019','p80p20_2020','p80p20_2021','p80p20_2022','p80p20_2023','p80p20_2024'],
                 var_name = 'Anio',
                 value_name = 'Desigualdad_p80p20')
p80p20.Anio = p80p20.Anio.str[-4:].astype('int64')
p80p20.Desigualdad_p80p20 = p80p20.Desigualdad_p80p20.fillna(p80p20.Desigualdad_p80p20.mean())
p80p20.rename(columns = {'Provincias':'Provincia'}, inplace = True)
info(p80p20)
p80p20.head()
```

```python
sociedades = pd.read_csv(ruta_sociedades, sep = ';')
info(sociedades)
sociedades.head()
```

```python
paro = pd.read_csv(ruta_paro, sep = ';')
paro.Provincia = paro.Provincia.str[3:]
info(paro)
paro.head()
```

```python
dataset = criminalidad.merge(inmigracion,
                             how = 'left',
                             left_on = ['Provincia','Trimestre','Anio'],
                             right_on =['Provincia','Trimestre','Año'])
dataset.drop(columns = ['Año'], inplace = True)
dataset.rename(columns = lambda x:f'Inm_{x}' if x in list(inmigracion.columns) and x not in list(criminalidad.columns) else x,
               inplace = True)

dataset = dataset.merge(edad_media,
                        how = 'left',
                        on = ['Provincia','Anio'])
dataset.rename(columns = lambda x:f'EM_{x}' if x in list(edad_media.columns) and x not in list(criminalidad.columns) else x,
               inplace = True)

dataset = dataset.merge(ipc,
                        how = 'left',
                        on = ['Provincia','Anio','Trimestre'])
dataset.rename(columns = lambda x:f'IPC_{x}' if x in list(ipc.columns) and x not in list(criminalidad.columns) else x,
               inplace = True)

dataset = dataset.merge(poblacion,
                        how = 'left',
                        on = ['Provincia','Anio','Trimestre'])
dataset.rename(columns = lambda x:f'Pobl_{x}' if x in list(poblacion.columns) and x not in list(criminalidad.columns) else x,
               inplace = True)

dataset = dataset.merge(gini,
                        how = 'left',
                        on = ['Provincia','Anio'])

dataset = dataset.merge(p80p20,
                        how = 'left',
                        on = ['Provincia','Anio'])

dataset = dataset.merge(sociedades,
                        how = 'left',
                        on = ['Provincia','Anio'])
dataset.rename(columns = lambda x:f'Soc_{x}' if x in list(sociedades.columns) and x not in list(criminalidad.columns) else x,
               inplace = True)

dataset = dataset.merge(paro,
                        how = 'left',
                        on = ['Provincia','Anio','Trimestre'])
dataset.rename(columns = lambda x:f'Paro_{x}' if x in list(paro.columns) and x not in list(criminalidad.columns) else x,
               inplace = True)

print(dataset.shape)
print(dataset.dtypes)
dataset.head()
```

```python
dataset.to_csv('./Transformados/Dataset_clustering.csv', sep = ';', index = False)
print(datetime.now(), ' - guardado')
```
